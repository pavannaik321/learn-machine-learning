```python
pie1 = np.array([22.2, 17.6, 8.8, 8, 7.7, 6.7])
myLable = ["Java", "Python", "PHP", "JavaScript", "C#", "C++"]
exploted = [0.2 , 0 ,0 , 0 ,0 ,0 ]
plt.pie(pie1, labels = myLable ,explode=exploted)
plt.show()
```
